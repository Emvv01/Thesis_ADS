#Cleaning the data

```python
# Convert the 'Antwoord' column to string type
scorius["Antwoord"] = scorius["Antwoord"].astype(str)

# Tokenize the text into individual words for each row
tokens = scorius['Antwoord'].apply(nltk.word_tokenize)

# Initializes the WordNet Lemmatizer from the NLTK library
lemmatizer = nltk.stem.WordNetLemmatizer()

# Lemmatize each word in the tokens for each row
scorius['lemmatized_words'] = tokens.apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Convert the lists of words to lowercase
scorius['cleanned_text'] = scorius['lemmatized_words'].apply(lambda x: [word.lower() for word in x])

# Removing digits
scorius['cleanned_text'] = scorius['cleanned_text'].apply(lambda x: [word for word in x if not word.isdigit()])

# Removing punctuations
scorius['cleanned_text'] = scorius['cleanned_text'].apply(lambda x: [word.translate(str.maketrans('', '', string.punctuation)) for word in x])

# Remove emojis using the emoji library
scorius['cleanned_text'] = scorius['cleanned_text'].apply(lambda x: [emoji.demojize(word, delimiters=(":", ": ")) for word in x])

# Remove emoji descriptions
scorius['cleanned_text'] = scorius['cleanned_text'].apply(lambda x: [word for word in x if not word.startswith(':')])

# Remove words with less than 3 letters
scorius['cleanned_text'] = scorius['cleanned_text'].apply(lambda x: [word for word in x if len(word) >= 3])
```

#LDA unigrams all word combinations

```python
# dataset for the unigrams all word combinations
uni_all = scorius

# Set the number of runs to obtain an average coherence score
runs = 10

# Create a dictionary
id2word = corpora.Dictionary(uni_all['cleanned_text'])

# Filter out words that occur in less than 30 documents or more than 60% of the documents
id2word.filter_extremes(no_below=5, no_above=0.2)

# Create a corpus
corpus = [id2word.doc2bow(doc) for doc in uni_all['cleanned_text']]

# Set the range of cluster numbers to evaluate
cluster_range = range(10, 30)

coherence_scores = []
avg_coherence_scores = []  # List to store average coherence scores
models = []

# Calculate coherence scores for each cluster number
for num_clusters in cluster_range:
    random.seed(42)
    model_coherences = []
    for _ in range(runs):
        model = LdaModel(corpus=corpus, num_topics=num_clusters)
        coherencemodel = CoherenceModel(model=model, texts=uni_all['cleanned_text'], dictionary=id2word, coherence='c_v')
        coherence_scores.append(coherencemodel.get_coherence())
        # Calculate coherence per run
        model_coherences.append(coherencemodel.get_coherence())
    # Calculate average coherence
    avg_coherence = sum(model_coherences) / runs
    # Append average coherence to separate list
    avg_coherence_scores.append(avg_coherence)
    models.append(model)

print(avg_coherence_scores)

# Visualize the coherence scores in a plot
plt.figure(figsize=(12, 8))
plt.plot(cluster_range, avg_coherence_scores, marker='o')
plt.xlabel("Number of topics")
plt.ylabel("Coherence Score")
plt.title("Coherence Scores for Different Numbers of topics")
plt.show()
```

```python
# Define the alpha and eta values to evaluate
alpha_range = [0.01, 0.05, 0.1, 0.5,1 ,5]
eta_range = [0.01, 0.05, 0.1, 0.5, 1, 5]

best_coherence = -1.0
best_alpha = None
best_eta = None

# Iterate over all combinations of alpha and eta
for alpha in alpha_range:
    for eta in eta_range:
        model = LdaModel(corpus=corpus, num_topics=10, id2word=id2word, alpha=alpha, eta=eta)
        coherence_model = CoherenceModel(model=model, texts=uni_all['cleanned_text'], dictionary=id2word, coherence='c_v')
        coherence_score = coherence_model.get_coherence()

        print("Alpha:", alpha, "Eta:", eta, "Coherence Score:", coherence_score)

        # Check if this is the best coherence score so far
        if coherence_score > best_coherence:
            best_coherence = coherence_score
            best_alpha = alpha
            best_eta = eta

# Print the best coherence score and corresponding alpha and eta values
print("Best Coherence Score:", best_coherence)
print("Best Alpha:", best_alpha)
print("Best Eta:", best_eta)
```

```python
# dataset for the bigrams all word combinations
bi_all = scorius['cleanned_text']

# Function to generate bigrams
def generate_bigrams(text):
    bigrams = []
    for i in range(len(text) - 1):

        # Filter out bigrams with the same words repeated
        if text[i] != text[i + 1]:
            bigram = text[i] + "_" + text[i + 1]
            bigrams.append(bigram)
    return bigrams

# Generate bigrams for all texts
all_bigrams = []
for text in bi_all:
    bigrams = generate_bigrams(text)
    all_bigrams.append(bigrams)

# Filter out texts that do not have any bigrams
all_bigrams = [bigrams for bigrams in all_bigrams if bigrams]

# Check if all_bigrams are empty
if not all_bigrams:
    print("No bigrams found. Please check your input texts.")
    exit()

```

```python
# Remove empty strings from the 'cleanned_text' column
all_bigrams = [bigrams for bigrams in all_bigrams if bigrams]

# Calculate the frequency of bigrams
bigram_frequency = {}

# Iterate over each list of bigrams
for bigrams in all_bigrams:
    # Iterate over each bigram in the list
    for bigram in bigrams:
        # Update the frequency count for the bigram
        if bigram in bigram_frequency:
            bigram_frequency[bigram] += 1
        else:
            bigram_frequency[bigram] = 1

# Create lists to store the bigrams and their frequencies
bigrams = []
frequencies = []

# Sort the bigram frequencies in descending order
sorted_bigram_frequency = sorted(bigram_frequency.items(), key=lambda x: x[1], reverse=True)

# Extract the top 20 sorted bigrams and frequencies into separate lists
top_20_bigrams = sorted_bigram_frequency[:20]
for bigram, frequency in top_20_bigrams:
    bigrams.append(bigram)
    frequencies.append(frequency)

# Increase the figure size
plt.figure(figsize=(12, 6))

# Create a bar chart for the top 20 bigrams
plt.bar(bigrams, frequencies)
plt.xlabel('Bigrams')
plt.ylabel('Frequency')
plt.title('Occurrence of the top 20 bigrams all word combinations')
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()

# Get the bigram with the highest frequency
most_frequent_bigram, highest_frequency = sorted_bigram_frequency[0]

# Print the bigram with the highest frequency
print("Bigram with the highest frequency:", most_frequent_bigram, highest_frequency)
```

```python
# Convert the bigrams back to sentences
all_bigrams = [' '.join(bigrams) for bigrams in all_bigrams]

# Perform TF-IDF calculations
tfidf_vectorizer = TfidfVectorizer(min_df=2)  # Adjust min_df as needed
tfidf_matrix = tfidf_vectorizer.fit_transform(all_bigrams)

# Check if the vocabulary is empty
if len(tfidf_vectorizer.vocabulary_) == 0:
    print("Empty vocabulary. Check your preprocessing steps and stopwords list.")
else:
    # Create a DataFrame from the TF-IDF matrix
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

    # Get the TF-IDF scores for the top 20 words
    top_20_words = tfidf_df.sum().nlargest(20)
    words = top_20_words.index.tolist()
    scores = top_20_words.values.tolist()

    # Create a bar plot for the top 20 words
    plt.figure(figsize=(12, 6))
    plt.bar(words, scores)
    plt.xlabel('Words')
    plt.ylabel('TF-IDF Score')
    plt.title('Top 20 bigrams by TF-IDF Score')
    plt.xticks(rotation='vertical')
    plt.tight_layout()
    plt.show()
```

```python
# Create a dictionary
id2word = corpora.Dictionary(all_bigrams)

# Filter out words that occur in less than 1 documents or more than 10% of the documents
id2word.filter_extremes(no_below=1, no_above=0.01)

# Create a corpus
corpus = [id2word.doc2bow(doc) for doc in all_bigrams]

# Set the number of runs to obtain an average coherence score
runs = 1

# Set the range of cluster numbers to evaluate
cluster_range = range(10, 30)

# Create list to store coherence score
coherence_scores = []

# Create list to store average coherence scores
avg_coherence_scores = []
models = []

# Calculate coherence scores for each cluster number
for num_clusters in cluster_range:
    random.seed(42)
    model_coherences = []
    for _ in range(runs):
        model = LdaModel(corpus=corpus, num_topics=num_clusters)
        coherencemodel = CoherenceModel(model=model, texts=all_bigrams, dictionary=id2word, coherence='c_v')
        coherence_scores.append(coherencemodel.get_coherence())

        # Calculate coherence per run
        model_coherences.append(coherencemodel.get_coherence())

    # Calculate average coherence
    avg_coherence = sum(model_coherences) / runs

    # Append average coherence to separate list
    avg_coherence_scores.append(avg_coherence)
    models.append(model)

# Print the average coherence score
print(avg_coherence_scores)

# Visualize the coherence scores in a plot
plt.figure(figsize=(12, 8))
plt.plot(cluster_range, avg_coherence_scores, marker='o')
plt.xlabel("Number of topics")
plt.ylabel("Coherence Score")
plt.title("Coherence Scores for Different Numbers of topics")
plt.show()
```

```python
# Define the alpha and eta values to evaluate
alpha_range = [0.01, 0.05, 0.1, 0.5,5]
eta_range = [0.01, 0.05, 0.1, 0.5,5]

best_coherence = -1.0
best_alpha = None
best_eta = None

# Iterate over all combinations of alpha and eta
for alpha in alpha_range:
    for eta in eta_range:
        model = LdaModel(corpus=corpus, num_topics=10, id2word=id2word, alpha=alpha, eta=eta)
        coherence_model = CoherenceModel(model=model, texts=all_bigrams, dictionary=id2word, coherence='c_v')
        coherence_score = coherence_model.get_coherence()

        print("Alpha:", alpha, "Eta:", eta, "Coherence Score:", coherence_score)

        # Check if this is the best coherence score so far
        if coherence_score > best_coherence:
            best_coherence = coherence_score
            best_alpha = alpha
            best_eta = eta

# Print the best coherence score and corresponding alpha and eta values
print("Best Coherence Score:", best_coherence)
print("Best Alpha:", best_alpha)
print("Best Eta:", best_eta)
```

```python
# Initialize the alpha and eta
alpha = 0.1
eta = 0.01

# Initialize the LDA model with specified parameters
model = LdaModel(corpus=corpus, num_topics=10,id2word=id2word, alpha=alpha, eta=eta)

# Calculate the coherence score of the model
coherence_model = CoherenceModel(model=model, texts=all_bigrams, dictionary=id2word, coherence='c_v')
coherence_score = coherence_model.get_coherence()

# Print the coherence score of the model
print("Coherence Score:", coherence_score)
```

```python
# Number op topic with the highest coherence value
N_TOPICS = 10

# Getting words with the highest probability for the each topic
for topic in range(N_TOPICS):
    words = model.show_topic(topic, 10)
    topic_n_words = ' '.join([word[0] for word in words])
    print('Topic {}: {}'.format(str(topic), topic_n_words))
```

```python
# Print the words with probabilities for each topic
model.print_topics()
```
